# Data Cleaning Example # 8: Temperature IoT

## Import

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Extract data from csv

```python
df = pd.read_csv('Datasets/IOT-temp.csv')
```

```python
df.head()
```

## Change to dateindex

```python
df.set_index(pd.DatetimeIndex(df['noted_date']), inplace=True)
```

```python
df.drop(['id', 'room_id/id', 'noted_date'], inplace=True, axis=1)
```

```python
df.head(10)
```

## Check for duplicates

```python
df = df.drop_duplicates()
```

```python
df.plot()
```

```python
df_in = df.loc[df['out/in'] == 'In']
```

```python
df_out = df.loc[df['out/in'] == 'Out']
```

```python
#Predicted vs True Adj Close Value – LSTM
plt.plot(df_in.temp)
plt.plot(df_out.temp)
plt.legend()
plt.show()
```
