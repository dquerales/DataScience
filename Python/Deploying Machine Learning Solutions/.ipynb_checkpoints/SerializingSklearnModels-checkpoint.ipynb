# Serializing Sk-learn Models

## Importing required libraries

```python
import pandas as pd
import numpy as np
```

## Loading the Data

```python
df = pd.read_csv('Datasets/CarPrice_Assignment.csv')
```

## Data Cleaning and Preparation

### Check the dimensions 

```python
df.shape
```

### Check columns names

```python
df.columns
```

```python
df.head(5)
```

### Check for Nulls

```python
df.isnull().values.any()
```

### Drop some columns

```python
df.drop(['car_ID', 'symboling', 'CarName'], axis=1, inplace=True)
```

```python
df.shape
```

### Get Categoricals

```python
df = pd.get_dummies(df)
```

```python
df.head()
```

## Machine Learning: LinearRegression

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
```

### Define Output and Inputs

```python
y = df['price']
X = df.drop(['price'], axis = 1)
```

### Split dataset

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
X_train.shape, X_test.shape
```

```python
y_train.shape, y_test.shape
```

### Instantiate Linear Regression Model

```python
model = LinearRegression().fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

### Score

```python
model.score(X_train, y_train)
```

```python
r2_score(y_test, y_pred)
```

## Serializing model

### Using JSON
is only good for simple models

```python
import json
```

#### Model coef & intercept 

```python
model.coef_
```

```python
model.intercept_
```

#### Save or Serializing model

```python
model_param = {}

model_param['coef'] = list(model.coef_)
model_param['intercept'] = model.intercept_.tolist()
```

```python
json_txt = json.dumps(model_param, indent=4)
```

```python
with open('Models/regressor_param.txt', 'w') as file:
    file.write(json_txt)
```

#### Load or deserializing models

```python
with open('Models/regressor_param.txt', 'r') as file:
    json_txt=json.load(file)
```

```python
json_model = LinearRegression()
```

```python
json_model.coef_ = np.array(json_txt['coef'])
json_model.intercept_ = np.array(json_txt['intercept'])
```

#### Test

```python
y_pred = json_model.predict(X_test)

r2_score(y_test, y_pred)
```

### Using Pickle
export like object

```python
import pickle
```

#### Save

```python
pickle.dump(model, open('Models/model.pkl', 'wb'))
```

#### Load

```python
pickle_model = pickle.load(open('Models/model.pkl', 'rb'))
```

#### Test

```python
y_pred = pickle_model.predict(X_test)

r2_score(y_test, y_pred)
```

### Using Joblib
is best for scikit-learn

```python
import joblib
```

```python
filename = 'Models/model.joblib'
```

#### Save

```python
joblib.dump(model, filename)
```

#### Load

```python
joblib_model = joblib.load(filename)
```

#### Test

```python
y_pred = joblib_model.predict(X_test)

r2_score(y_test, y_pred)
```
