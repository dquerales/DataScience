## Machine Learning  model

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

### Splitting X and y into training and testing sets

```python
X_train, X_test, y_train, y_test = train_test_split(df[['Adj Close']], 
                                                    df[['EMA_10']], 
                                                    test_size=.2, 
                                                    random_state=1)
```

### Create linear regression object

```python
model = LinearRegression()
```

### Train the model using the training sets

```python
model.fit(X_train, y_train)
```

### Regression coefficients

```python
print('Coefficients: ', model.coef_)
```

### Model Score

```python
model.score(X_test, y_test)
```

### Sort dates for Plots

```python
y_test = y_test.sort_index()
X_test = X_test.sort_index()
```

### Evaluate test values

```python
y_pred = model.predict(X_test)
```

### Plotting 

```python
#Predicted vs True Adj Close Value – LSTM
plt.plot(y_test, label='True Value')
plt.plot(X_test.index, y_pred, label='Pred Value')
plt.title('Prediction by LR')
plt.xlabel('Date')
plt.ylabel('USD [$]')
plt.legend()
plt.show()
```

At this point, we’ve trained a model on historical pricing data using the Adjusted Closing value and the Exponential Moving Average for a 10-day trading period. Our goal was to develop a model that can use the EMA of any given day (dependent on pricing from the previous 9 days) and accurately predict that day’s closing price. Let’s run a simulation of a very simple trading strategy to assess how well we might have done using this.

Strategy: If our model predicts a higher closing value than the opening value we make a trade for a single share on that day—buying at market open and selling just before market close.

```python
y_pred = model.predict(df[['Adj Close']])
```

```python
df_pred = pd.DataFrame(y_pred, columns = ['Prediction'], index = df.index.copy())
```

```python
df2 = pd.read_csv('Datasets/TSLA.csv')
df2.set_index(pd.DatetimeIndex(df2['Date']), inplace=True)
```

```python
df_pred = pd.concat([df2[['Open']], df[['Adj Close']], df_pred], axis=1)
```

```python
df_pred['Trade'] = df_pred['Prediction'] - df_pred['Open'] > 0
```
