### Change data types

```python
df['total_cases'] = pd.to_numeric(df['total_cases'])
df['total_deaths'] = pd.to_numeric(df['total_deaths'])
df['total_vaccinations'] = pd.to_numeric(df['total_vaccinations'])
df['date'] = pd.to_datetime(df['date'])
```

```python
max_cases = df.groupby(by='location').max().sort_values('total_cases')
max_cases
```

```python
plt.figure(figsize=(10,8))
plt.bar(max_cases.index, max_cases['total_cases'])

plt.title('Sales by Months')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.show()
```
